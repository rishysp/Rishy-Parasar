```python
gb= [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
str1="Player 1 is winner"
str2="Player 2 is winner"
def display_gameboard(gb):
    
    for i in gb:
        print(i)
    
def check_if_empty(x,y):
    if gb[x][y]==' ':
        return True
    else:
        return False
def player1_choice(a,b):
    if check_if_empty(a,b):
        gb[a][b]='X'
        return 1
    else:
        print("It's filled")
        return 0
def player2_choice(a,b):
    if check_if_empty(a,b):
        gb[a][b]='O'
        return 1
    else:
        return 0
def check_for_winner():
    if gb[0][0]==gb[0][1]==gb[0][2]=='X' or gb[1][0]==gb[1][1]==gb[1][2]=='X'  or gb[2][0]==gb[2][1]==gb[2][2]=='X' or gb[0][0]==gb[1][1]==gb[2][2]=='X' or gb[0][2]==gb[1][1]==gb[2][0]=='X':
        return str1
    elif  gb[0][0]==gb[0][1]==gb[0][2]=='O' or gb[1][0]==gb[1][1]==gb[1][2]=='O'  or gb[2][0]==gb[2][1]==gb[2][2]=='O' or gb[0][0]==gb[1][1]==gb[2][2]=='O' or gb[0][2]==gb[1][1]==gb[2][0]=='O':
        return str2
    else:
        return ' '
    
def if_board_full():
    if gb[0][0]==' ' or gb[0][1]==' ' or gb[0][2]==' ' or gb[1][0]==' ' or gb[1][1]==' ' or gb[1][2]==' ' or gb[2][0]==' ' or gb[2][1]==' ' or gb[2][2]==' ' :
        return True
    else:
        return False
print("Begin Game!")
while if_board_full():
    print("Player 1 enter the value of x & y coordinates:")
    x, y = map(int, input().split())
    player1_choice(x,y)
    display_gameboard(gb)
    res=check_for_winner()
    if res == ' ':
        pass
    else:
        print(res)
        break
    print("Player 2 enter the value of x & y coordinates:")
    x, y = map(int, input().split())
    player2_choice(x,y)
    display_gameboard(gb)
    
    res=check_for_winner()
    if res == ' ':
        continue
    else:
        print(res)
        break
if res==' ':
    print("Draw")

    
```
